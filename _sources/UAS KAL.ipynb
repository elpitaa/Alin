#  1. Hitunglah PageRank dari graph berikut
image.png

```python
import numpy as np

# Adjacency matrix of the graph
adjacency_matrix = np.array([
    [0, 1, 0, 0, 0],  # Node 1 -> Node 2
    [0, 0, 1, 0, 1],  # Node 2 -> Node 3, Node 5
    [1, 1, 0, 0, 1],  # Node 3 -> Node 1, Node 2, Node 5
    [1, 1, 0, 0, 1],  # Node 4 -> Node 1, Node 2, Node 5
    [0, 1, 0, 1, 0],  # Node 5 -> Node 2, Node 4
], dtype=float)

# Number of nodes
n = adjacency_matrix.shape[0]

# Damping factor
d = 0.85

# Initialize the PageRank vector with equal values
pagerank = np.ones(n) / n

# Number of iterations
iterations = 100

# Power Iteration method
for _ in range(iterations):
    new_pagerank = np.zeros(n)
    for j in range(n):
        for k in range(n):
            if adjacency_matrix[k, j] != 0:
                new_pagerank[j] += d * pagerank[k] / np.sum(adjacency_matrix[k])
    new_pagerank += (1 - d) / n
    pagerank = new_pagerank

print("PageRank values:")
for i, rank in enumerate(pagerank):
    print(f"Node {i+1}: {rank:.4f}")
```

#  2.	Carilah penyelesaian dari Sistem persamaan berikut  dengan Eliminasi Gauss

$$
\begin{align*}
x_1 - x_2 + x_3 &= 3 \\
2x_1 + x_2 + 3x_3 &= 8 \\
6x_1 + x_2 - x_3 &= 4
\end{align*}
$$



```python
import numpy as np

# Definisikan matriks augmented
A = np.array([[1, -1, 1, 3],
              [2, 1, 3, 8],
              [6, 0, -1, 4]], dtype=float)

# Lakukan eliminasi Gauss
n = A.shape[0]

for i in range(n):
    # Pivoting
    max_row = np.argmax(np.abs(A[i:, i])) + i
    A[[i, max_row]] = A[[max_row, i]]

    # Buat elemen di bawah pivot menjadi 0
    for j in range(i+1, n):
        factor = A[j, i] / A[i, i]
        A[j, i:] -= factor * A[i, i:]

# Back-substitution
x = np.zeros(n)
for i in range(n-1, -1, -1):
    x[i] = (A[i, -1] - np.sum(A[i, i+1:n] * x[i+1:n])) / A[i, i]

print("Solusi dari sistem persamaan adalah:")
print(x)
```
