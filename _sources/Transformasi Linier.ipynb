# Transformasi
Transformasi matriks adalah konsep dalam aljabar linear di mana matriks digunakan untuk mewakili transformasi linear dalam ruang vektor. Ini sering digunakan dalam bidang seperti grafika komputer, pemrosesan sinyal, dan analisis data.

```python
import numpy as np

# Definisikan matriks transformasi
A = np.array([[2, 0],
              [0, 3]])

# Definisikan vektor yang akan ditransformasikan
x = np.array([1, 2])

# Lakukan transformasi linier
y = np.dot(A, x)

print("Vektor asal:", x)
print("Vektor setelah transformasi:", y)
```

# 1. Refleksi
Refleksi mencerminkan objek terhadap suatu garis, bidang, atau titik tertentu. Dalam transformasi matriks 2x2, refleksi terhadap sumbu x, sumbu y, atau garis ùë¶=ùë• dilakukan dengan menggunakan matriks refleksi yang sesuai.

```python
import numpy as np

# Matriks refleksi terhadap sumbu y
Fy = np.array([[-1, 0],
               [0, 1]])

# Contoh koordinat objek sebelum refleksi
koordinat_sebelum = np.array([[2],
                              [3]])

# Melakukan refleksi terhadap sumbu y
koordinat_setelah = np.dot(Fy, koordinat_sebelum)

print("Koordinat objek sebelum refleksi:")
print(koordinat_sebelum)

print("\nKoordinat objek setelah refleksi:")
print(koordinat_setelah)
```

# Refleksi Terhadap Sumbu X dan Y

```python
import numpy as np

def refleksi_terhadap_sumbu_x(koordinat):
    # Matriks refleksi terhadap sumbu x
    Fx = np.array([[1, 0],
                   [0, -1]])

    # Melakukan refleksi terhadap sumbu x
    koordinat_setelah = np.dot(Fx, koordinat)
    return koordinat_setelah

def refleksi_terhadap_sumbu_y(koordinat):
    # Matriks refleksi terhadap sumbu y
    Fy = np.array([[-1, 0],
                   [0, 1]])

    # Melakukan refleksi terhadap sumbu y
    koordinat_setelah = np.dot(Fy, koordinat)
    return koordinat_setelah

# Contoh koordinat objek
koordinat_objek = np.array([[2],
                             [3]])

# Refleksi terhadap sumbu x
hasil_refleksi_x = refleksi_terhadap_sumbu_x(koordinat_objek)

# Refleksi terhadap sumbu y
hasil_refleksi_y = refleksi_terhadap_sumbu_y(koordinat_objek)

print("Koordinat objek sebelum refleksi:")
print(koordinat_objek)

print("\nKoordinat objek setelah refleksi terhadap sumbu x:")
print(hasil_refleksi_x)

print("\nKoordinat objek setelah refleksi terhadap sumbu y:")
print(hasil_refleksi_y)
```

# 2. Shearing
Transformasi shearing adalah jenis transformasi geometri di mana titik-titik suatu objek digeser sepanjang sumbu tertentu, tanpa mengubah ukuran objek tersebut.

```python
def shearing_transform(points, shear_factor):
    """
    Melakukan transformasi shearing pada sebuah objek berdasarkan faktor shearing yang diberikan.

    Args:
    - points: List dari tuple koordinat titik-titik objek.
    - shear_factor: Tuple berisi faktor shearing untuk sumbu x dan sumbu y.

    Returns:
    - List dari tuple koordinat titik-titik objek setelah transformasi shearing.
    """
    return [(x + shear_factor[0] * y, y + shear_factor[1] * x) for x, y in points]

# Contoh penggunaan
if __name__ == "__main__":
    # Definisikan titik-titik objek awal (dalam format list dari tuple)
    points = [(0, 0), (1, 0), (1, 1), (0, 1)]

    # Faktor shearing (misalnya, shearing pada sumbu x sebesar 0.5 dan pada sumbu y sebesar 0.3)
    shear_factor = (0.5, 0.3)

    # Melakukan transformasi shearing
    transformed_points = shearing_transform(points, shear_factor)

    # Menampilkan hasil transformasi
    print("Titik-titik setelah transformasi shearing:")
    for point in transformed_points:
        print(point)
```

# 3. Rotasi
Transformasi rotasi adalah jenis transformasi geometri di mana suatu objek diputar sekitar titik tertentu yang disebut pusat rotasi. Dalam rotasi, setiap titik objek diputar sekitar pusat rotasi sebesar sudut tertentu dalam arah tertentu. kode Python untuk melakukan rotasi dua dimensi terhadap suatu titik (x, y) sebesar sudut tertentu (dalam radian) searah jarum jam:

```python
import math

def rotate_point(x, y, angle):
    # Mengubah sudut rotasi menjadi radian
    angle_rad = math.radians(angle)

    # Menghitung koordinat baru setelah rotasi
    x_rotated = x * math.cos(angle_rad) - y * math.sin(angle_rad)
    y_rotated = x * math.sin(angle_rad) + y * math.cos(angle_rad)

    return x_rotated, y_rotated

# Contoh penggunaan
x = 1
y = 1
angle = 45  # Sudut rotasi dalam derajat

x_rotated, y_rotated = rotate_point(x, y, angle)

print("Koordinat setelah rotasi:", x_rotated, y_rotated)
```

# Rotasi 90 Derajat

```python
import numpy as np

# Membuat matriks 2x2 dengan nilai awal
matriks = np.array([[1, 0], [0, 1]])

# Mendefinisikan matriks rotasi untuk rotasi 90 derajat searah jarum jam
theta = np.radians(90)  # Mengubah derajat menjadi radian
rotasi_matriks = np.array([[np.cos(theta), -np.sin(theta)],
                           [np.sin(theta), np.cos(theta)]])

# Melakukan rotasi matriks
matriks_rotasi = np.dot(rotasi_matriks, matriks)

# Menampilkan matriks setelah rotasi
print("Matriks setelah rotasi 90 derajat searah jarum jam:")
print(matriks_rotasi.round())
```

# 4. Squeezing
Transformasi squeezing adalah jenis transformasi geometri di mana objek dipersempit atau diperlebar dalam satu atau lebih arah. Hal ini dilakukan dengan memperbesar atau memperkecil jarak antara titik-titik objek sepanjang sumbu tertentu.

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk melakukan squeezing horizontal
def squeeze_horizontal(points, factor):
    return points * np.array([factor, 1])

# Fungsi untuk melakukan squeezing vertikal
def squeeze_vertical(points, factor):
    return points * np.array([1, factor])

# Titik-titik objek awal
points = np.array([[1, 1], [2, 2], [3, 1]])

# Faktor squeezing horizontal dan vertikal
horizontal_factor = 0.5
vertical_factor = 1.5

# Lakukan transformasi squeezing horizontal dan vertikal
squeezed_horizontal_points = squeeze_horizontal(points, horizontal_factor)
squeezed_vertical_points = squeeze_vertical(points, vertical_factor)

# Plot hasil transformasi
plt.scatter(points[:, 0], points[:, 1], label='Original Points')
plt.scatter(squeezed_horizontal_points[:, 0], squeezed_horizontal_points[:, 1], label='Squeezed Horizontal Points')
plt.scatter(squeezed_vertical_points[:, 0], squeezed_vertical_points[:, 1], label='Squeezed Vertical Points')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Squeezing Transformation')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```

# 5. Stretching
Transformasi stretching adalah jenis transformasi geometri di mana objek diperlebar atau dipersempit dalam satu atau lebih arah, dengan memperbesar atau memperkecil jarak antara titik-titik objek sepanjang sumbu tertentu.

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk melakukan stretching horizontal
def stretch_horizontal(points, factor):
    return points * np.array([factor, 1])

# Fungsi untuk melakukan stretching vertikal
def stretch_vertical(points, factor):
    return points * np.array([1, factor])

# Titik-titik objek awal
points = np.array([[1, 1], [2, 2], [3, 1]])

# Faktor stretching horizontal dan vertikal
horizontal_factor = 2
vertical_factor = 0.5

# Lakukan transformasi stretching horizontal dan vertikal
stretched_horizontal_points = stretch_horizontal(points, horizontal_factor)
stretched_vertical_points = stretch_vertical(points, vertical_factor)

# Plot hasil transformasi
plt.scatter(points[:, 0], points[:, 1], label='Original Points')
plt.scatter(stretched_horizontal_points[:, 0], stretched_horizontal_points[:, 1], label='Stretched Horizontal Points')
plt.scatter(stretched_vertical_points[:, 0], stretched_vertical_points[:, 1], label='Stretched Vertical Points')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Stretching Transformation')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
