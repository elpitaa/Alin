```python
import numpy as np

# Membuat matriks 2x2
A = np.array([[1, 4], [2, 3]])

# Menghitung eigenvalues dan eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

# Menampilkan hasil
print("Matriks A:")
print(A)
print("\nEigenvalues:")
print(eigenvalues.round())
print("\nEigenvectors:")
print(eigenvectors.round())
```
