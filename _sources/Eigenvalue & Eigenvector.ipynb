# Eigenvalue  & Eigenvector
Eigenvalue dan eigenvector adalah konsep penting dalam aljabar linear yang memiliki banyak aplikasi dalam berbagai bidang, termasuk fisika, statistik, dan pembelajaran mesin.
*  Eigenvalue
Eigenvalue adalah suatu nilai skalar
𝜆yang memenuhi persamaan: 𝐴𝑣=𝜆𝑣
*  Eigenvector adalah suatu vektor bukan nol
𝑣 yang, ketika dikalikan dengan matriks
𝐴 menghasilkan suatu vektor yang merupakan skalar kali dari
𝑣 itu sendiri. Dengan kata lain,
𝑣 tidak berubah arah dalam transformasi yang diberikan oleh matriks
𝐴 hanya skala (panjang) nya saja yang berubah.




```python
import numpy as np

# Membuat matriks 2x2
A = np.array([[1, 4], [2, 3]])

# Menghitung eigenvalues dan eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

# Menampilkan hasil
print("Matriks A:")
print(A)
print("\nEigenvalues:")
print(eigenvalues.round())
print("\nEigenvectors:")
print(eigenvectors.round())
```
