# sifat-sifat Determinant

1.  hanya dilakukan oleh matriks
2.  determinan dari identitas= 1
3.  jika suatu matriks dikalikan dengan konstanta maka rumus determinanny:
det(kA)=k^n x det A
4.


```python
import numpy as np


#fungsi mencari determinan
def det(matriks):
    return np.linalg.det(matriks)

print("Membuktikan bahwa determinan dari matriks identitas adalah 1")
I=np.array([[1,0,0],[0,1,0],[0,0,1]])
print('Matriks A =\n', I)
print("determinan dari matriks i adalah",det(I))

print("perubahan baris pada matriks identitas untuk mencari determinannya")
I2=np.array([[1,0,0],[0,0,1],[0,1,0]])
print("determinan dari matriks i setelah diubah pertama kali adalah",det(I2))
I3=np.array([[0,1,0],[0,0,1],[1,0,0]])
print("determinan dari matriks i setelah diubah kedua kali adalah",det(I3),"\n")

print("mencari determinan c")
c = np.array([[1,2,1],[3,1,2],[2,1,2]])
print("determinan dari matriks c adalah",det(c))
d = np.array([[3,1,2],[1,2,1],[2,1,2]])
print("determinan dari matriks c setelah diubah pertama kali adalah",det(d))
e = np.array([[1,2,1],[2,4,2],[2,1,2]])
print("determinan dari matriks e adalah",det(e))
print("jika baris dikalikan dengan kelipatan maka hasilnya akan 0")
f = np.array([[1,2,1],[2,4,2],[2,4,3]])
print("determinan dari matriks f adalah",det(f))
g = np.array([[6,2,4],[2,4,2],[4,2,4]])
print("determinan dari matriks g adalah",det(g))#2^3(3)
a=np.array([[3,4],[2,1]])
print("determinan dari matriks a adalah",det(a))
h=np.array([[6,8],[4,2]])
print("determinan dari matriks h adalah",det(h)) #2^2(-5)=-20
i=np.array([[6,8],[2,1]])
print("determinan dari matriks i adalah",det(i)) #jika hanya satu baris saja c det(a) = 2(-5) = -10

print("\nmengalikan salah satu kolom ")
j=np.array([[6,4],[4,1]])
print("determinan dari matriks j adalah",det(j))#hasilnya bakal sama dengan yang salah satu baris saja yang dikalikan

#penumlahan dari determinan A dan (hasil dari A1 dan A2)
g = np.array([[4,1],[2,1]])
print("determinan dari matriks g adalah",det(g))
g = np.array([[7,5],[2,1]])
print("determinan dari matriks g adalah",det(g))
```
